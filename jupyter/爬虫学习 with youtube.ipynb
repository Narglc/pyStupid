```python
#coding:utf-8
from bs4 import BeautifulSoup
html_sample = '\
<html> \
  <body> \
    <h1 id="title">Hello World</h1> \
    <a href="#" class="link">This is link1</a> \
    <a href="# link2" class="link">This is link2</a>\
  <body>\
<html>'

soup = BeautifulSoup(html_sample,"html5lib")
print "pure text:"+soup.text       
print soup.contents  
print "class:"+soup.select('.link')[0].text
print "id:"+soup.select('#title')[0].text
print "by tabs:"+soup.select('html')[0].text.strip()
```

```python
#天涯博客文章爬去
import requests
from bs4 import BeautifulSoup
res = requests.get("http://blog.tianya.cn/blog-773339-1.shtml")
soup=BeautifulSoup(res.text,'html5lib')

for article in soup.select('.article'):
    print "Title:"+article.select('a')[0].text+'     Time:'+article.select('.gray6')[0].text.strip()+"  Contents:"+article.select('p')[0].text
```
